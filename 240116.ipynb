```python
class Calculator:
    def __init__(self, value):
        self.value = value

    def add(self, other):
        self.value += other
        return self

    def subtract(self, other):
        self.value -= other
        return self

    def multiply(self, other):
        self.value *= other
        return self

    def get_value(self):
        return self.value

calc = Calculator(1)
result = calc.add(2).subtract(1).multiply(3).get_value()
# result = calc.add(2).subtract(1).multiply(3).get_value()    # get_value 의 결과가 int.

print(result)  # 결과: 6
```

```python
# 우리가 실제 코딩하는 것은 여러 클래스 또는 인스턴스의 조합

class Operator:
    def add(self, x, y):
        return x + y

    def sub(self, x, y):
        return x - y

    def mul(self, x, y):
        return x * y

    def div(self, x, y):
        return x / y

    def _and(self, x, y):
        pass

    def _or(self, x, y):
        pass

op = Operator()
logical_op = {
    'add': op.add,
    'sub': op.sub,
    'mul': op.mul,
    'div': op.div,
}
arithmetic_op = {
    '_and': op._and,
    '_or': op._or,
}

#**************************
print(logical_op['add'](2, 3)) # 출력: 5
#**************************

print(op.add(2, 3)) # 출력: 5

for _, f in logical_op.items():
    print(f(2, 3)) # 4칙연산 모두 계산
```

```python
# 함수를 리턴할 수 있다 (데코레이터의 핵심 개념 )
# x 라는 변수가 함수가 끝났는데도 살아 있다(adder 에서 x 값) ==> 클로저
def create_adder(x):
    def adder(y):
        return x + y
    return adder

add_5 = create_adder(5)
print(add_5(10))  # 출력: 15
```

```python
# 함수를 리턴할 수 있다. ==> 데코레이터
def creat_sq(x):
    def sq(y):
        return x**y
    return sq

제곱2 = creat_sq(2)
제곱3 = creat_sq(2)
제곱4 = creat_sq(2)

제곱2(2), 제곱2(3), 제곱2(4)    # 여러개 반환하면 반환값이 튜플 --> unpacking

a, b, c = 제곱2(2), 제곱2(3), 제곱2(4)

print(a, b, c)
```

```python
def add_and_subtract(x, y):
    addition = x + y
    subtraction = x - y
    return addition, subtraction

# 함수 호출
result = add_and_subtract(5, 3)

# 결과 출력
print(result)  # 출력: (8, 2)
```

## 재귀함수 (내가 나를 호출하는 함수)
* 정말 자신있지 않으면 반복문으로 사용
* 비효율 최고
* 무한반복될 수 있음 (실제 재귀는 1000번 반복)
* 복잡도가 높아질수록 실수 가능성 높아짐

```python
# f(5) --> 5 * 4 * 3 * 2 * 1 == 120 == 5!    # factorial
# 분할 정복, 다이나믹프로그래밍

def f(n):
    if n <= 1:
        return n
    return n * f(n - 1)

f(5)
```

```python
def 숫자출력(count):
    if count > 100:
        return      # None
    print(count)
    return 숫자출력(count+1) #값을 1부터 반복횟수 까지의 값을 출력


숫자출력(1)
```

```python
# 설명
def f(n):
    if n <= 1:
        return n
    return n * f(n-1)

# f(5)        5 * f(4) == 5 * 24 => 120
# f(4)        4 * f(3) == 4 * 6
# f(3)        3 * f(2) == 3 * 2
# f(2)        2 * f(1) == 2 * 1
# f(1)        1 (return n)
```
